```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

# Knockouts 

### Read in data

```python
df_od = pd.read_csv("20251015_KOs_OD.csv",index_col=0)
df_pos = pd.read_csv("20251015_KOs_pos.csv",index_col=0)
strains = df_pos['strain'].unique()
```

### HM9 ONLY

```python
growth_rates = []

for i in strains: 

    # Define the exponential-phase window
    start_tp = 1.5
    end_tp   = 2.5
    
    # get replicate row positions for this strain
    positions = np.array(df_pos.loc[(df_pos['strain']==i) & (df_pos['salt']==0)].index)
    strain_od = df_od.loc[positions]
    
    # restrict to exponential region
    time = df_od.T.index.values.astype(float)
    mask = (time >= start_tp) & (time <= end_tp)
    time_exp = time[mask]
    
    # log-transform OD (avoid log(0))
    od_log = np.log(strain_od + 1e-9)
    
    # Fit slope = growth rate    
    for j in range(len(positions)):
        od_exp = od_log.iloc[j][mask]
        lr = LinearRegression().fit(time_exp.reshape(-1,1), od_exp)
        mu = lr.coef_[0]  # per unit time
        growth_rates.append({"strain": i, "growth_rate": mu})

# Convert results to DataFrame
growth_df = pd.DataFrame(growth_rates)

# -------------------------------
#            Plot
# -------------------------------

plt.figure(figsize=(6,4))
sns.boxplot(data=growth_df, x='strain', y='growth_rate', palette='Set2')
plt.title("HM9")
plt.xlabel("Strain")
plt.ylabel("Growth rate (1/time)")
plt.ylim(0.3,1)
```
