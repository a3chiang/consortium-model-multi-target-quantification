```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Rectangle
import matplotlib
from scipy.signal import savgol_filter 
import re
```

# Define colormap 

```python
# ------------------------------------------------------------------

# Custom nonlinear colormap (white â†’ green)
def NonLinCdict(steps, hexcol_array):
    cdict = {'red': (), 'green': (), 'blue': ()}
    for s, hexcol in zip(steps, hexcol_array):
        rgb = matplotlib.colors.hex2color(hexcol)
        cdict['red'] += ((s, rgb[0], rgb[0]),)
        cdict['green'] += ((s, rgb[1], rgb[1]),)
        cdict['blue'] += ((s, rgb[2], rgb[2]),)
    return cdict

hc = ['#ffffff', '#53c653', '#339933', '#206020', '#0d260d']
th = [0, 0.45, 0.6, 0.8, 1]

cdict = NonLinCdict(th, hc)
cm = matplotlib.colors.LinearSegmentedColormap('test', cdict)

# ------------------------------------------------------------------
```

# Tap Water

```python
# Load and reorder fold-change data
tap_data = pd.read_csv('./tap_foldchange.csv', index_col=0)
desired_order = ['arsR', 'zntA', 'merR', 'zraP', 'cusC', 'cadC']
tap_data = tap_data.reindex(desired_order)

# Initialize figure and axes
fig, ax = plt.subplots(figsize=(5, 5))

# Remove default axis ticks and spines
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

# Manually position heatmap axis
ax = fig.add_axes([0.18, 0.3, 0.6, 0.55])

# Draw heatmap (log2 fold change)
ax = sns.heatmap(
    data=np.log2(tap_data),
    vmin=0,
    vmax=7,
    cmap=cm,
    linewidths=1,
    cbar=False
)

# Add outer border
ax.add_patch(Rectangle(
    (0, 0),
    tap_data.shape[1] - 0.1,
    tap_data.shape[0] - 0.01,
    fill=False,
    edgecolor='black',
    lw=1
))

# Add inner border
ax.add_patch(Rectangle((0, 0), tap_data.shape[1]-1, tap_data.shape[0]-0.01, fill=False, edgecolor='black', lw=1))


# Axis labels and formatting
xticks = ['As', 'Cd', 'Hg', 'Pb', 'Cu', 'NT']

ax.set_yticklabels(desired_order, fontsize=25, rotation=0)
ax.set_xticklabels(xticks, fontsize=25, rotation=90)

ax.tick_params(axis='both', which='both', length=0)
ax.set_title("Tap", fontsize=25)

plt.show()
```

# Autofluorescence - media only with no cells added

```python
# Load autofluorescence data and transpose so conditions are columns
autoFL_df = pd.read_csv('./autoFL.csv', index_col=0).T

# Initialize results DataFrame
autoFL = pd.DataFrame(columns=['water', 'autoGFP', 'color'])

# Color mapping for each condition
color_map = {
    'hm9': 'red',
    'tap': 'orange',
    'sea': 'blue',
    'treatment': 'purple',
    'apple': 'green'
}

# Conditions to analyze
conditions = ['hm9', 'tap', 'sea', 'apple', 'treatment']

# Compute mean autofluorescence for each condition
for condition in conditions:
    mean_auto = autoFL_df[condition].mean().mean()
    autoFL.loc[len(autoFL)] = [condition, mean_auto, color_map[condition]]

# Plot results
plt.title('Mean Autofluorescence', size=12)
plt.bar(autoFL['water'], autoFL['autoGFP'], color=autoFL['color'])
```
