```python
# Import packages 

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import ast
from scipy.stats import linregress
```

# Dosage effects on GR

### Read in CSVs

```python
# GFP fold-change 
GR_df = pd.read_csv('GR.csv',index_col=0)

# Sensors and their cognate heavy metal 
metal_sensor = pd.read_csv('sensors.csv')
metal_sensor = metal_sensor.sort_values(by='sensor')
```

### arsR to As 

```python
slopes_df = pd.DataFrame(columns=['ppm','hm','sensor','GR','R2'])

hm = 'As'
sensor='arsR'

# Input HM + sensor to get all concentrations 
plot_df = GR_df.loc[(GR_df.hm==hm) & (GR_df.sensor==sensor)]

# Sort slopes by increasing concentration
plot_df = plot_df.sort_values(by='ppm')

# Plotting 
plt.figure(figsize=(4, 3))
plt.bar(plot_df['ppm'].astype(str), plot_df['GR'], width=0.6, color='skyblue', edgecolor='black')

# Add horizontal line at a specific value (e.g., 0.25)
plt.axhline(y=plot_df.loc[plot_df.ppm == plot_df.ppm.min()]['GR'].values[0], color='coral', linestyle='--', linewidth=2, label='0ppm')
plt.legend(loc='upper left', bbox_to_anchor=(0, 1.02), ncol=1)

# Labels and title
plt.xlabel('Concentration (ppm)',size=10)
plt.ylabel('Growth Rate (h$^{-1}$)',size=10)
plt.title('arsR Growth with Arsenic Induction',size=10)
plt.xticks(fontsize=8,rotation=45)
plt.yticks(fontsize=8)
plt.tight_layout()
plt.show()
```
